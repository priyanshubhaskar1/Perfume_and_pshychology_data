```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
import requests

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'
}

webpage = requests.get('https://pmc.ncbi.nlm.nih.gov/articles/PMC5198031/#ref-list1', headers=headers).text

```

```python
soup = BeautifulSoup(webpage, 'lxml')
```

```python
print(soup.prettify())
```

```python
table=soup.find_all('table')[0]
```

```python
print(table)
```

```python
world_title= table.find_all('th')
```

```python
world_title
```

```python
world_table_titles = [title.text for title in world_title]
print(world_table_titles)
```

```python
df = pd.DataFrame(columns = world_table_titles)
df
```

```python
column_data=table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]

    lenght= len(df)
    df.loc[lenght] = individual_row_data


    
```
